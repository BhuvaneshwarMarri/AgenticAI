```python
from dotenv import load_dotenv
load_dotenv()

import os
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain_core.messages import AIMessage, HumanMessage
from pprint import pprint

messages = [AIMessage(content=f"Please tell me how can I help",name="LLModel")]
messages.append(HumanMessage(content=f"I want to learn coding",name="Smaari"))
messages.append(AIMessage(content=f"Which programming language you want to learn", name="LLModel"))
messages.append(HumanMessage(content=f"I want to learn java",name="Smaari"))

for message in messages:
    message.pretty_print()
```

```python
from langchain_groq import ChatGroq
llm = ChatGroq(model="openai/gpt-oss-120b")
result = llm.invoke(messages)
```
